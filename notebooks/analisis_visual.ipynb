```python
# Importar librerías 
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Configuración inicial de seaborn
sns.set_context('paper')
sns.set_style('whitegrid')
```

```python
# Cargar base de datos
encuesta = pd.read_csv('../datos/procesados/encuesta_filtrada.csv')
```

```python
# Visualizar columnas
print("Columnas de la base de datos:")
for columna in encuesta.columns:
    print(columna)
```

# Variables cuantitativas

```python
# Histograma edad
sns.histplot(data=encuesta, x='edad', hue='genero', bins=22, kde=True, palette='magma')

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de las edades de los encuestados')
ticks = list(range(22,45))
plt.xticks(ticks)

ruta_guardado = '../reportes/figuras/histograma_edad.png'
plt.savefig(ruta_guardado)

plt.show()
```

```python
# Conteo edad
sns.countplot(data = encuesta, x = 'edad')

plt.title('Distribución de las edades de los encuestados:')
plt.xlabel('Edad')
plt.ylabel('Cantidad')

plt.show()
```

```python
# Histograma experiencia
sns.histplot(data=encuesta, x='experiencia', kde=True, bins=20)

plt.xlabel('Años de experiencia')
plt.ylabel('Frecuencia')
plt.title('Distribución de los años de experiencia de los encuestados')

ruta_guardado = '../reportes/figuras/histograma_experiencia.png'
plt.savefig(ruta_guardado)

plt.show()
```

```python
# Histograma experiencia
sns.countplot(data = encuesta, x = 'experiencia')

plt.xlabel('Años de experiencia')
plt.ylabel('Frecuencia')
plt.title('Distribución de los años de experiencia de los encuestados')

plt.show()
```

```python
# Histograma salario
sns.histplot(data=encuesta, x='salario', kde=True, hue='genero', palette='magma', bins=16)

plt.xlabel('Salario en Euros')
plt.ylabel('Frecuencia')
plt.title('Distribución de los salarios de los encuestados')

ruta_guardado = '../reportes/figuras/histograma_salario.png'
plt.savefig(ruta_guardado)

plt.show()
```
